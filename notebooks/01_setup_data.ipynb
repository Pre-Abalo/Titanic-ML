# 01 — Mise en place & chargement des données

**Objectif** : démarrer proprement, vérifier l'environnement, charger les CSV, et faire des contrôles rapides de qualité.

**À la fin** : vous aurez les données `train.csv` et `test.csv` chargées et prêtes pour l'exploration.

## 1) Imports et vérification d'environnement

```python
import sys, os
import numpy as np
import pandas as pd

print(f"Python: {sys.version.split()[0]}")
print("NumPy:", np.__version__)
print("Pandas:", pd.__version__)
```

## 2) Chemins et chargement des fichiers

```python
from pathlib import Path

DATA_DIR = Path("../data")
TRAIN_PATH = DATA_DIR / "train.csv"
TEST_PATH  = DATA_DIR / "test.csv"

if not TRAIN_PATH.exists() or not TEST_PATH.exists():
    raise FileNotFoundError(
        "Placez `train.csv` et `test.csv` dans le dossier `data/`.\n" 
        "Téléchargement: https://www.kaggle.com/c/titanic "
    )

train = pd.read_csv(TRAIN_PATH)
test  = pd.read_csv(TEST_PATH)

print(train.shape, test.shape)
train.head(3)
```

## 3) Contrôles rapides

```python
# Types, valeurs manquantes, aperçu statistique
display(train.info())
display(train.describe(include='all'))
train.isna().mean().sort_values(ascending=False).head(10)
```

```python
# Types, valeurs manquantes, aperçu statistique
display(train.info())

# Affichage des statistiques descriptives (avec correction)
display(train.describe(include='all'))

# Optionnel : analyse spécifique pour les colonnes datetime
datetime_columns = train.select_dtypes(include=['datetime64', 'datetime64[ns]'])
if not datetime_columns.empty:
    display(datetime_columns.describe())

# Identifier les colonnes avec des valeurs manquantes (top 10)
train.isna().mean().sort_values(ascending=False).head(10)
```

## 4) Rappel des colonnes principales

- **Survived** *(cible, dans `train` uniquement)* : 0 = non, 1 = oui  
- **Pclass** : classe du billet (1 = 1ère, 2, 3)  
- **Sex** : sexe  
- **Age** : âge  
- **SibSp** : # de frères/soeurs + conjoint à bord  
- **Parch** : # de parents + enfants à bord  
- **Fare** : prix du billet  
- **Embarked** : port d'embarquement (C, Q, S)

## 5) Sauvegarde d'un échantillon pour suivi

```python
sample = train.sample(5, random_state=42)
sample.to_csv("../outputs/sample_rows.csv", index=False)
sample
```
